```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import joblib

# Load dataset
df = pd.read_csv("dataset.csv")

# Fill NaNs with 'none'
df.fillna('none', inplace=True)

# Combine all symptoms into one list and encode them
symptom_columns = [col for col in df.columns if 'Symptom' in col]
symptoms = sorted(set(symptom for col in symptom_columns for symptom in df[col].unique()))
symptoms = [s for s in symptoms if s != 'none']

# Create binary columns for each symptom
for symptom in symptoms:
    df[symptom] = df[symptom_columns].apply(lambda x: int(symptom in x.values), axis=1)

# Drop original Symptom columns
X = df.drop(columns=symptom_columns + ['Disease'])
y = df['Disease']

# Encode the target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train the best model: Random Forest
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
preds = model.predict(X_test)
acc = accuracy_score(y_test, preds)
print(f"Random Forest Accuracy: {acc:.4f}")
print(classification_report(y_test, preds, target_names=le.classes_))

# Save the model and label encoder
joblib.dump(model, "disease_model.pkl")
joblib.dump(le, "label_encoder.pkl")

print("Model and label encoder saved.")
```
